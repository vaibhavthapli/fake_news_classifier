# concate  df and y

```python
test = pd.concat([df,y], axis =1)
```

```python
test = test[["title","label"]]
```

```python
test
```

```python
test = test.dropna()
```

```python
test
```

```python
X = test["title"]
y = test["label"]
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer_obj = Tokenizer()
tokenizer_obj.fit_on_texts(X)
test_sam = tokenizer_obj.texts_to_sequences(X)
test_sam_pad = pad_sequences(test_sam,maxlen = 30)

```

```python
# load json and create model
from keras.models import model_from_json
json_file = open('model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("best_model.best.hdf5")
print("Loaded model from disk")
```

```python
a = loaded_model.predict_classes(test_sam_pad)
```

```python
a
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(classification_report(a,y))
```

```python
y
```

```python
y_pred = loaded_model.predict_classes(test_sam_pad, batch_size = 128)

true = 0
for i, y in enumerate(y):
    if y == y_pred[i]:
        true += 1

print('Correct Prediction: {}'.format(true))
print('Wrong Prediction: {}'.format(len(y_pred) - true))
print('Accuracy: {}'.format(true/len(y_pred)*100))
```
